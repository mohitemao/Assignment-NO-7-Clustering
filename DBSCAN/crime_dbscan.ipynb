```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
crime=pd.read_csv('C:/Users/hp/Desktop/DATA SCIENCE COURSE/ASSIGNMENT/ASSIGNMENT 7/crime_data.csv')
print(crime.head())
```

```python
crime1=crime.drop(crime.iloc[:,0:1],axis=1,)
print(crime1.head())
```

```python
crime1.info()
```

```python
scaler=StandardScaler()
scaler.fit(crime1)
crime_scale=scaler.fit_transform(crime1)
crime_scale
```

```python
dbscan=DBSCAN(eps=1.3, min_samples=6)
dbscan.fit(crime_scale)
```

```python
dbscan.labels_
```

```python
cluster=pd.DataFrame(dbscan.labels_,columns=['clusterid'])
```

```python
cluster
```

```python
crime_new=pd.concat([crime,cluster],axis=1)
```

```python
crime_new.head()
```

```python
crime_new.sort_values('clusterid').head()
```

```python
crime_new.groupby('clusterid').agg('mean').reset_index()
```
