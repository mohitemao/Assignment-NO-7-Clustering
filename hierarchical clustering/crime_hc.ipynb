```python
# normalizing data
def norm_function(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
# normalise data 
crime_norm=norm_function(crime.iloc[:,1:])
crime_norm.head()
```

```python
# create dendrogram
plt.figure(figsize=(10,7))
dendrogram=sch.dendrogram(sch.linkage(crime_norm, method='complete'))
# "complete " linkage method is the distance between the most distant elements from each class
```

```python
# create cluster
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hc
```

```python
y_hc=hc.fit_predict(crime_norm)
clusters=pd.DataFrame(y_hc, columns=['clusters'])
clusters
```

```python
crime_norm['cluster_id']=clusters
```

```python
clusters
```

```python
crime_norm.sort_values('cluster_id').head(50)
```

```python
crime_norm.groupby('cluster_id').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10,7))
plt.scatter(crime_norm['cluster_id'],crime_norm['Murder'],c=y_hc)
```

```python
final_data=pd.concat([crime,clusters],axis=1)
final_data
```
